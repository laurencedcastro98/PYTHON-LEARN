#Condicionales sin parámetro y conversión de tipo de dato

Al realizar una función con condicionales pero sin parámetro, al generar el resultado, da error por el simple hecho de que el valor de la edad al no estar con el parámetro lo lee como si el numero fuese un texto. En otras palabras es como un string (str) por lo que hay que pasarlo a entero(int), para este caso se utilizar **int** para definir el valor sin parámetro.

```python
def verificar_si_puede_conducir_sin_parametros():
  edad = input('Digite su edad: ')
  edad = int(edad)
  if edad >= 18:
    print("Usted tiene edad suficiente para conducir")
  else:
    print("Usted NO tiene edad suficiente para conducir")

```

#Bucles y ciclos (Loops)

**En esta parte se realiza un bucle fuera de la función que nos permita detectar por edades quien esta apto para conducir, como se ve abajo se hace la función, y despues se crea al ciclo o bucle, recordando que las palabras *FOR* y *IN* es lo mismo que *Para cada* y *Dentro* solo que en escritura inglesa, por eso es importante memorizar esto.** *Ver abajo*

```python
#Bucle fuera de la función
edades = [18, 15, 12, 45, 50]

def verificar_si_puede_conducir(edad):
   if edad >= 18:
    print("Usted tiene edad suficiente para conducir")
    
   else:
     print("Usted NO tiene edad suficiente para conducir")

#para cada (for), dentro(in) es necesario en lo que sea requerido con bucles o ciclos
for edad in edades:
  verificar_si_puede_conducir(edad)

```

```python
#Bucle dentro de la función
edades = [18, 15, 12, 45, 50]

def verificar_si_puede_conducir_con_bucle(edades):
  for edad in edades:
   if edad >= 18:
    print("Usted tiene edad suficiente para conducir")
    
   else:
     print("Usted NO tiene edad suficiente para conducir")


verificar_si_puede_conducir_con_bucle(edades)
```

#Tipo Booleano

Variables pueden ser atribuidos por dos variables, true o false, 1 o 0. solo dos valores dos caminos

```python
edad = 18
```

```python
edad >= 18
```

```python
edad < 18
```

```python
edad == 18
```

```python
verificaciones = []
edades = [13, 15, 20]

def verificar_si_puede_conducir_bool (verificaciones, edades):
  for edad in edades:
    if edad >= 18:
      verificaciones.append(True)
    else:
      verificaciones.append(False)
  for verificacion in verificaciones:
    if verificacion == True:
      print('Usted tiene edad suficiente para conducir')
    else:
      print('Usted NO tiene edad suficiente para conducir')


verificar_si_puede_conducir_bool(verificaciones, edades)

```
